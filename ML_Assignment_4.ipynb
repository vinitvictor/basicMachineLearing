```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from DecisionTree import TreeNode,DecisionTree
```

```python
trainDataset=pd.read_csv("train.csv")
testDataset=pd.read_csv("test.csv")
trainDataset.head()
```

```python
trainDataset=trainDataset.drop(["Opponent","Date","ID"], axis=1)
testDataset=testDataset.drop(["Opponent","Date","ID"], axis=1)
trainDataset.head()
```

```python
#Categorical data => numerical trainDataset
for col in trainDataset.columns:
    trainDataset[col]=pd.Categorical(trainDataset[col])
    trainDataset[col]=trainDataset[col].cat.codes
    
trainDataset
```

```python
#Categorical data => numerical testDataset
for col in testDataset.columns:
    testDataset[col]=pd.Categorical(testDataset[col])
    testDataset[col]=testDataset[col].cat.codes
testDataset
```

```python
#spearting labels and features
X= trainDataset[trainDataset.columns[:-1]].values
Y=trainDataset['Label'].values
X
```

```python
predFeatureX=testDataset[testDataset.columns[:-1]].values
xGroundTruth=testDataset["Label"].values
predFeatureX
```

```python
xGroundTruth
```

```python
#Naive baiyes prediction

model=GaussianNB()
model.fit(np.array(X),Y)
predY=model.predict(predFeatureX)
```

```python
# Calculation of accuracy and fscore

def accuracyCheck(name, y, y_pred):
    headers = {
        'Accuracy': accuracy_score, 
        'F1_Score': f1_score, 
        'Precision': precision_score, 
        'Recall': recall_score
    }
    
    with open('{}.csv'.format(name), 'w') as summary:
        print(','.join(headers.keys()), file=summary)
        summary.write(','.join(['{:3.3f}'.format(metric(y, y_pred)) for metric in headers.values()]))
```
